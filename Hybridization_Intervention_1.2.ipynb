## <center>7. Orbitals and Hybridization

---
---

## <div style = 'background-color:gainsboro'> <center>Orbitals</div>

<div style = 'background-color:gainsboro'>Due to electron's wave-particle duality, we can describe the discrete energy level where the electron can exist in an atom with its wavefunction 𝛹. Since wavefunctions are a function of spatial location, we can square the wavefunction (𝛹<sup>2</sup>) to show the probability of finding the electron in said location, also refered to as an electron cloud. To visualize this probability distribution we can generate an image of the atomic orbitals by plotting a three-dimensional plot of 𝛹<sup>2</sup>

>Organic chemistry primarily deals with first-and second-row elements (H, C, N, O), so we can focus on the 1s, 2s, and three 2p (2p<sub>x</sub>, 2p<sub>y</sub>, 2p<sub>z</sub>) orbitals.<br>

>Chemists use an energy diagram to represent relative energy levels of said atomic orbitals.<br>
<br>
Degenerate orbitals are orbitals with the same energy level, like the three 2p orbitals.</div>

# <div style = 'background-color:gainsboro'> <center>Hybridization</div>

<div style = 'background-color:gainsboro'> Valence Bond Theory, the theory that an electron from each atom are shared to form a covalent bond, was produced before the theory of quantum mechanics was established, and thus was incomplete. Hybridization was developed to incorporate quantum mechanics to help explain the shapes of molecules that were observed from experimentation.<br>
Hybridization is a mathematical averaging of atomic orbitals s and p, which allowed chemists to understand bond strength, bond rotation, bond geometry, resonance and chemical magnetism. In this activity, we will be focusing on bond geometry to qualitatively understand electron behavior.<br>  
Students often struggle to understand hybridization because they unproductively apply Valence Shell Electron Pair Repulsion (VSEPR) theory to determine electron structure. VSEPR theory does not account for the stabilizing quantum mechanical effects of orbital overlap, and overly relies on electron-pair repulsion.</div>
